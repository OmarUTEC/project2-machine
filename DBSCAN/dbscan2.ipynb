```python
import pandas as pd
from sklearn.decomposition import PCA

# Cargar dataset_tissue.txt
data_path = '../db/dataset_tissue.txt'
data = pd.read_csv(data_path, sep=',', header=0, index_col=0)

# Cargar clase.txt
classes_path = '../db/clase.txt'
classes = pd.read_csv(classes_path, sep=',', header=0, index_col=0)

# Transponer los datos
data_transposed = data.transpose()


# Aplicar PCA para retener el 95% de la varianza
pca_95 = PCA(n_components=0.95)
data_pca = pca_95.fit_transform(data_transposed)

print('Número de componentes principales: ', pca_95.n_components_)
```

## DBSCAN

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from collections import Counter

# Implementación de DBSCAN
def distance(v1, v2):
    return np.linalg.norm(v1 - v2)

def dbscan(data, epsilon, min_samples):
    cluster_id = 0
    labels = np.zeros(len(data))
    visited = np.zeros(len(data), dtype=bool)

    def range_query(point, epsilon):
        return np.where(np.array([distance(point, x) for x in data]) < epsilon)[0]

    def expand_cluster(p, neighbors, cluster_id):
        labels[p] = cluster_id
        i = 0
        while i < len(neighbors):
            q = neighbors[i]
            if not visited[q]:
                visited[q] = True
                q_neighbors = range_query(data[q], epsilon)
                if len(q_neighbors) >= min_samples:
                    neighbors = np.concatenate((neighbors, q_neighbors))
            if labels[q] == 0:
                labels[q] = cluster_id
            i += 1

    for i in range(len(data)):
        if visited[i]:
            continue
        visited[i] = True
        neighbors = range_query(data[i], epsilon)
        if len(neighbors) < min_samples:
            labels[i] = -1  # Marcar como ruido (outlier)
        else:
            cluster_id += 1
            expand_cluster(i, neighbors, cluster_id)
    
    return labels

# Ajusta estos valores según tus necesidades
epsilon = 50
min_samples = 1
cluster_labels = dbscan(data_pca, epsilon, min_samples)

# Obtener clústeres únicos (excepto los ruidos)
unique_clusters = set(cluster_labels) - {-1}

# Obtener nombres de tejidos
tejidos = pd.read_csv('../db/clase.txt', header=0)
tejidos_dict = dict(zip(tejidos.index + 1, tejidos['x']))  # Mapeo de números de cluster a nombres de tejidos

# Crear el gráfico de dispersión 2D y etiquetar cada punto
plt.figure(figsize=(12, 10))
scatter = plt.scatter(data_pca[:, 0], data_pca[:, 1], c=cluster_labels, cmap='jet', marker='o', edgecolor='k', s=50, alpha=0.6)

for cluster_id in unique_clusters:
    cluster_center = data_pca[cluster_labels == cluster_id].mean(axis=0)
    plt.text(cluster_center[0], cluster_center[1], tejidos_dict[cluster_id], fontsize=10, ha='right', va='bottom')

plt.colorbar(scatter)
plt.title('Clustering DBSCAN con Nombres de Tejidos')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score, homogeneity_score, completeness_score

# Calcular métricas de calidad del clustering
silhouette_avg = silhouette_score(data_pca, cluster_labels)
davies_bouldin = davies_bouldin_score(data_pca, cluster_labels)
calinski_harabasz = calinski_harabasz_score(data_pca, cluster_labels)
completeness = completeness_score(cluster_labels, tejidos.index + 1)
# Calcular Homogeneity
homogeneity = homogeneity_score(tejidos['x'], cluster_labels)
print(f"Homogeneity: {homogeneity}")

print(f"Silhouette Score: {silhouette_avg}")
print(f"Davies-Bouldin Score: {davies_bouldin}")
print(f"Calinski-Harabasz Score: {calinski_harabasz}")
print(f"Completeness Score: {completeness}")
```
